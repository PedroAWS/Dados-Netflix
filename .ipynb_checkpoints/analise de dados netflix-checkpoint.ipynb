```python
import pandas as pd 
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px 
%matplotlib inline

import seaborn as sns
sns.set(style='darkgrid', font_scale=1.2, color_codes=True)

#filtrando apenas o tipo `Movie`
#tabela.drop(tabela.loc[tabela['Tipo']=='TV Show'].index, inplace=True)
tabela = pd.read_csv("netflix_titles.csv")

# transformar tabela em portugues 
tabela = tabela.rename({
    'type': 'Tipo',
    'description': 'Descrição',
    'director': 'Diretor',
    'cast': 'Elenco',
    'title': 'Titulo',
    'country': 'País',
    'date_added': 'Data_add',
    'release_year': 'Lançamento',
    'rating': 'Avaliação',
    'duration': 'Duração',
    'listed_in': 'Listado_em',
    'show_id': 'Exposição',
    }, axis = 1)
```

```python
# 2 visualizar a base de dados  e excluir a coluna de `Descrição` 

# 3 tratando dados VAZIOS da tabela 
# DELETANDO COLUNAS VAZIAS 
# DELETANDO LINHAS VAZIAS 
tabela = tabela.dropna(how="any" , axis=0)

df = tabela.copy()
display(tabela)



```

```python
#Principais paises que fazem filme País

countries={}
netflix_movies['País']=netflix_movies['País'].fillna('Unknown')
cou=list(netflix_movies['País'])
for i in cou:
    #print(i)
    i=list(i.split(','))
    if len(i)==1:
        if i in list(countries.keys()):
            countries[i]+=1
        else:
            countries[i[0]]=1
    else:
        for j in i:
            if j in list(countries.keys()):
                countries[j]+=1
            else:
                countries[j]=1
countries_fin={}
for country,no in countries.items():
    country=country.replace(' ','')
    if country in list(countries_fin.keys()):
        countries_fin[country]+=no
    else:
        countries_fin[country]=no
        
countries_fin={k: v for k, v in sorted(countries_fin.items(), key=lambda item: item[1], reverse= True)}

plt.figure(figsize=(8,8))
ax = sns.barplot(x=list(countries_fin.keys())[0:10],y=list(countries_fin.values())[0:10])
ax.set_xticklabels(list(countries_fin.keys())[0:10],rotation = 90)
```
